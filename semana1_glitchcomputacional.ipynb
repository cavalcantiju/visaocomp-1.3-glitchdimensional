```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Nome do arquivo
caminho_arquivo = "/content/img.jpg"

try:
    # Abre a imagem e converte para escala de cinza
    img = Image.open(caminho_arquivo).convert('L')
    imagem_original = np.array(img)

    # 1. Flatten (Vetor 1D)
    vetor_1d = imagem_original.flatten()

    # 2. Reshape "Glitch" (256x1024)
    imagem_glitch = vetor_1d.reshape(256, 1024)

    # Visualização
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(imagem_original, cmap='gray')

    plt.subplot(1, 2, 2)
    plt.title("Glitch Dimensional")
    plt.imshow(imagem_glitch, cmap='gray')

    plt.show()

except FileNotFoundError:
    print("Erro: O arquivo não foi encontrado. Você subiu a imagem na pasta lateral?")
```

O que aprendemos com esse Glitch:

A Memória é Linear: Na RAM, não existe quadrado. Os pixels estão um do lado do outro. Quando a gente faz o reshape para 256x1024 (em vez de 512x512), a gente forçou o computador a "quebrar a linha" no lugar errado.

O efeito de "Fatiamento": Perceberam que a imagem pareceu cortada e repetida? Isso acontece porque, como a largura dobrou (de 512 para 1024), cada linha da nova imagem agora contém exatamente duas linhas da imagem original colocadas lado a lado.

Transposta vs. Rotação: * A Transposta troca as linhas pelas colunas. Se a memória era lida como "anda X, pula uma linha", agora ela é lida como "anda Y, pula uma coluna".

Isso é diferente de rotacionar. Na rotação, você move os dados de lugar. Na transposta (especialmente no NumPy), você muitas vezes só muda a ordem de leitura (os strides), o que é muito mais rápido.
